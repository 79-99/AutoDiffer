```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

params = {'axes.labelsize': 20,
          'axes.titlesize': 30,
          "font.family": "serif",
          "font.size":12
         }
plt.rcParams.update(params)
```

```python
def plot_data(df, name):
    f, axarr = plt.subplots(ncols=2, figsize=(10,4))
    axarr[0].plot(df['equations'], df['single'], 
                  marker='o', linewidth=0.3, color='k', 
                  fillstyle='none', label='single')
    axarr[0].plot(df['equations'], df['multi'], 
                  marker='s', linewidth=0.3, color='k', 
                  fillstyle='none', label='multi')

    axarr[1].plot(df['equations'], df['single']/df['multi'], 
                  marker='^', linewidth=0.3, 
                  color='k', fillstyle='none')

    for ax in axarr:
        ax.set_xlabel("Number of equations")
    axarr[0].set_ylabel("Time (milliseconds)")
    axarr[1].set_ylabel("Improvement")

    axarr[0].legend()

    axarr[1].plot()
    f.subplots_adjust(wspace = 0.3)

    f.suptitle(name, size=25)
    plt.show()
```

```python
plot_data(pd.read_csv('std_case_A.txt', names=["equations", "single", "multi"]), "Case A: std::thread")
```

```python
plot_data(pd.read_csv('openmp_case_A.txt', names=["equations", "single", "multi"]), "Case A: openmp")
```

```python
plot_data(pd.read_csv('std_case_B.txt', names=["equations", "single", "multi"]), "Case B: std::thread")
```
