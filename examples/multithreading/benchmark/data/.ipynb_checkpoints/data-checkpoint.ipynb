```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

params = {'axes.labelsize': 20,
          'axes.titlesize': 30,
          "font.family": "serif",
          "font.size":12
         }
plt.rcParams.update(params)
```

```python
data = pd.read_csv('std_case_A.txt', names=["equations", "single", "multi"])
data.head()
```

```python
f, axarr = plt.subplots(ncols=2, figsize=(10,4))
axarr[0].plot(data['equations'], data['single'], 
              marker='o', linewidth=0.3, color='k', 
              fillstyle='none', label='single')
axarr[0].plot(data['equations'], data['multi'], 
              marker='s', linewidth=0.3, color='k', 
              fillstyle='none', label='multi')

axarr[1].plot(data['equations'], data['single']/data['multi'], 
              marker='^', linewidth=0.3, 
              color='k', fillstyle='none')

for ax in axarr:
    ax.set_xlabel("Number of equations")
axarr[0].set_ylabel("Time (milliseconds)")
axarr[1].set_ylabel("Improvement")

axarr[0].legend()

axarr[1].plot()
f.subplots_adjust(wspace = 0.3)

f.suptitle('Case A: std::thread', size=)
plt.show()
```

```python
data = pd.read_csv('data_openmp.txt', names=["threads", "single", "multi"])
data.head()
```
