```python
import matplotlib.pyplot as plt
from matplotlib import ticker
import numpy as np
from scipy.special import binom
from scipy.stats import norm
from matplotlib.pyplot import cm
from tqdm import tqdm
import pandas as pd

params = {'axes.labelsize': 20,
          'axes.titlesize': 30,
          "font.family": "serif",
          "font.size":12
         }
plt.rcParams.update(params)
```

```python
data = pd.read_csv('data.txt', names=["threads", "single", "multi"])
data.head()
```

```python
f, axarr = plt.subplots(ncols=2, figsize=(10,4))
axarr[0].plot(data['threads'], data['single'], marker='o', linewidth=0.3, color='k', fillstyle='none', label='single')
axarr[0].plot(data['threads'], data['multi'], marker='s', linewidth=0.3, color='k', fillstyle='none', label='multi')

axarr[1].plot(data['threads'], data['single']/data['multi'], marker='^', linewidth=0.3, 
              color='k', fillstyle='none')

for ax in axarr:
    ax.set_xlabel("Number of equations")
axarr[0].set_ylabel("Time (milliseconds)")
axarr[1].set_ylabel("Improvement")

axarr[0].legend()

axarr[1].plot()
f.subplots_adjust(wspace = 0.3)
plt.show()
```
